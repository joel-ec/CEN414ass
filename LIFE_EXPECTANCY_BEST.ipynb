```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from statistics import mean, variance, stdev, mode, median
import pandas 
```

```python
df_london = pd.read_csv(r'C:\Users\Joel\Downloads\Country_data.csv')





df_london = df_london.sort_values('2009')
df_london
```

```python
df_london = df_london.rename(columns = {' Country':'Country'})
```

```python
df_london.set_index('Country', inplace=True)
```

```python
for year in range(2010,2015):
    fig,ax = plt.subplots(1)
    df_london.plot.bar(ax = ax, y=str(year),figsize=(50,15), color='g', title=f"A GRAPH OF LIFE EXPECTANCY AT BIRTH OF COUNTRIES AGAINST THE COUNTRIES {year} ")
    
    ax.set_xlabel("COUNTRIES")
    ax.set_ylabel("NUMBER OF LIFE EXPECTANCY AT BIRTH")
```

```python
iqrs =[]
for year in range(2010,2015):
    iqrs.append(df_london[str(year)].quantile(0.75) - df_london[str(year)].quantile(0.25))
    
    
    
plt.ylabel("Interquatile range")
plt.xlabel("Years")
plt.title("INTERQUATILE RANGE OF LIFE EXPECTANCY")
plt.plot(range(2010,2015),iqrs)
```

```python
col = ['2010','2011','2012','2013','2014','2015']
_range = df_london[col].max() - df_london[col].min()
_range.plot.line()
plt.ylabel("Range")
plt.xlabel("Years")
plt.title("RANGE OF LIFE EXPECTANCY")
```

```python
plt.bar(col,df_london[col].mean())
plt.xlabel('years')
plt.ylabel('mean')
plt.title('MEAN OF LIFE EXPECTANCY')
```

```python
plt.bar(col,df_london[col].median())
plt.xlabel('years')
plt.ylabel('median')
plt.title('MEDIAN OF LIFE EXPECTANCY')
```

```python
plt.bar(col,df_london[col].std())
plt.xlabel('years')
plt.ylabel('standard deviation')
plt.title('STANDARD DEVIATION OF LIFE EXPECTANCY')
```

```python
plt.bar(col,df_london[col].var())
plt.xlabel('years')
plt.ylabel('Variance')
plt.title('VARIANCE OF LIFE EXPECTANCY')
```
